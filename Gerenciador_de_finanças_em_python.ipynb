# Código para gestão financeira em python

###Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python

class GestaoFinanceira:
    def __init__(self):
        self.df = pd.DataFrame(columns=['Data', 'Descricao', 'Valor', 'Tipo'])

    def adicionar_transacao(self, data, descricao, valor, tipo):
        nova_transacao = pd.Series([data, descricao, valor, tipo], index=self.df.columns)
        self.df = self.df.append(nova_transacao, ignore_index=True)

    def calcular_saldo(self):
        saldo = self.df.loc[self.df['Tipo'] == 'Receita', 'Valor'].sum() - self.df.loc[self.df['Tipo'] == 'Despesa', 'Valor'].sum()
        return saldo

    def gerar_relatorio_financeiro(self):
        relatorio = self.df.groupby(['Tipo'])['Valor'].sum()
        return relatorio

    def visualizar_dados_grafico(self):
        plt.figure(figsize=(8, 6))
        plt.bar(self.df['Data'], self.df['Valor'], color=self.df['Tipo'].map({'Receita': 'green', 'Despesa': 'red'}))
        plt.xlabel('Data')
        plt.ylabel('Valor')
        plt.title('Visualização Financeira')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    def definir_meta_financeira(self, meta):
        saldo_atual = self.calcular_saldo()
        if meta <= saldo_atual:
            return "Parabéns! Você atingiu a meta financeira."
        else:
            return "Você ainda precisa economizar mais para atingir a meta financeira."

# Exemplo de uso
gestao = GestaoFinanceira()
gestao.adicionar_transacao('2023-06-15', 'Salário', 15000.00, 'Receita')
gestao.adicionar_transacao('2023-06-20', 'Aluguel', 3800.00, 'Despesa')
gestao.adicionar_transacao('2023-06-25', 'Compras', 5000.00, 'Despesa')
gestao.adicionar_transacao('2023-06-28', 'Bônus', 2000.00, 'Receita')

saldo_atual = gestao.calcular_saldo()
print("Saldo Atual: R$%.2f" % saldo_atual)

relatorio_financeiro = gestao.gerar_relatorio_financeiro()
print("Relatório Financeiro:")
print(relatorio_financeiro)

gestao.visualizar_dados_grafico()

meta_financeira = 8000.00
mensagem_meta = gestao.definir_meta_financeira(meta_financeira)
print(mensagem_meta)
```
