```python
import matplotlib.pyplot as plt
import pandas as pd

# Dados de vendas mensais (substitua isso pelos seus próprios dados)
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro','Novembro','Dezembro']
vendas = [200000, 320000, 250000, 230000, 240000, 260000, 310000, 290000, 300000, 220000, 250000, 470000]

# Cálculos de métricas
media_vendas = sum(vendas) / len(vendas)
soma_vendas = sum(vendas)
var_percentual = [(vendas[i] - vendas[i - 1]) / vendas[i - 1] * 100 if i > 0 else 0 for i in range(len(vendas))]
tendencia_crescimento = (vendas[-1] - vendas[0]) / vendas[0] * 100
maior_venda_mensal = max(vendas)

# Cálculo da média móvel simples (SMA) de 3 meses
sma_3meses = [sum(vendas[i:i+3]) / 3 if i + 3 <= len(vendas) else sum(vendas[i:]) / (len(vendas) - i) for i in range(len(vendas))]

# Criação de um gráfico de barras para variação percentual
plt.figure(figsize=(12, 6))
plt.bar(meses[1:], var_percentual[1:], color='g', label='Variação Percentual em relação ao Mês Anterior')
plt.xlabel('Mês')
plt.ylabel('Variação Percentual')
plt.title('Variação Percentual Mensal em relação ao Mês Anterior')
plt.legend()
plt.grid(True)

```

```python
# Criação do gráfico de linha para vendas mensais
plt.figure(figsize=(12, 6))
plt.plot(meses, vendas, marker='o', linestyle='-', color='b', label='Vendas Mensais')
plt.plot(meses, sma_3meses, linestyle='--', color='m', label='Média Móvel de 3 Meses')
plt.axhline(y=media_vendas, color='r', linestyle='--', label='Média de Vendas')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.title('Vendas Mensais e Métricas')
plt.legend()
plt.grid(True)

# Exibição dos gráficos
plt.show()

# Exibição das métricas
print(f'A média de vendas mensais é de {media_vendas:.2f}')
print(f'A soma total das vendas é de {soma_vendas}')
print(f'A tendência de crescimento ao longo dos meses é de {tendencia_crescimento:.2f}%')
print(f'A maior venda mensal foi de {maior_venda_mensal}')
```

```python


# Cálculos de métricas
var_percentual = [(vendas[i] - vendas[i - 1]) / vendas[i - 1] * 100 if i > 0 else 0 for i in range(len(vendas))]

# Cálculo da distribuição da variação percentual
crescimento = sum(1 for var in var_percentual if var > 0)
decrescimento = sum(1 for var in var_percentual if var < 0)
estabilidade = len(var_percentual) - crescimento - decrescimento

# Criação do gráfico de pizza para explicar o KPI
labels = ['Crescimento', 'Decréscimo', 'Estabilidade']
sizes = [crescimento, decrescimento, estabilidade]
colors = ['g', 'r', 'y']
explode = (0.1, 0.1, 0.1)  # Explodir as fatias

plt.figure(figsize=(10, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Distribuição da Variação Percentual em relação ao Mês Anterior')
plt.axis('equal')  # Garante que o gráfico de pizza seja circular

# Exibição do gráfico de pizza
plt.show()

# Exibição das métricas
print(f'Número de meses com crescimento: {crescimento}')
print(f'Número de meses com decréscimo: {decrescimento}')
print(f'Número de meses de estabilidade: {estabilidade}')
# Exibição das métricas
print(f'A média de vendas mensais é de {media_vendas:.2f}')
print(f'A soma total das vendas é de {soma_vendas}')
print(f'A tendência de crescimento ao longo dos meses é de {tendencia_crescimento:.2f}%')
print(f'A maior venda mensal foi de {maior_venda_mensal}')
```
